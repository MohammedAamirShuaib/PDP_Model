```python
def get_url(url):

    try:

        if len(url) <50:
            pdp_check = "skipped"
        elif any(x in url for x in ['google', 'bing', 'wikipedia','yahoo', 'facebook', 'twitter', 'youtube']):
            pdp_check = "others"
        else:
            headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
            r = requests.get(url, headers=headers, timeout=20)
            soup = BeautifulSoup(r.content, 'lxml')
            title = soup.find_all(text=True)
            new = []
            for i in title:
                if i == None:
                    pdp_check = "no_URL"
                else:
                    new.append(i.lower())
                    add_to_cart = [i for i in new if i== 'add to cart']
                    #buy_now = [i for i in new if "buy now" in i]
                    if len(add_to_cart)>0 and len(add_to_cart)<10 :
                        pdp_check = "pdp"
                    else:
                        pdp_check = "not_pdp"
    except:
        pdp_check = "error"
    f = open('pdp_check.txt', 'a+')
    f.write(pdp_check+"\n")
    f.close()
    return pdp_check
```

```python
for i in data.URL:
    pdp_check = get_url(i)
    f = open('pdp_check.txt', 'a+')
    f.write(pdp_check+"\n")
    f.close()
    
```
